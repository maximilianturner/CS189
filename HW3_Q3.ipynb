```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

mean = np.array([1, 1])

cov = np.array([[1, 0], [0, 2]])

x, y = np.mgrid[-2:4:.01, -2:4:.01]

pos = np.dstack((x, y))

multi_norm = scipy.stats.multivariate_normal(mean, cov)

plt.contourf(x, y, multi_norm.pdf(pos))
plt.show() 
```

```python
mean1 = np.array([0, 2])

cov1 = np.array([[2, 1], [1, 1]])

mean2 = np.array([2, 0])

cov2 = np.array([[2, 1], [1, 1]])

x, y = np.mgrid[-3:5:.01, -2:4:.01]

pos = np.dstack((x, y))

multi_norm1 = scipy.stats.multivariate_normal(mean1, cov1)
multi_norm2 = scipy.stats.multivariate_normal(mean2, cov2)

plt.contourf(x, y, multi_norm1.pdf(pos) - multi_norm2.pdf(pos))
plt.show()
```
